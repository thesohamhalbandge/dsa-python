## Assignment Operators

````python
x = 10
y = 20

print(x + y) # Addition
print(x - y) # Subtraction
print(x * y) # Multiplication
print(x / y) # Division
print(x // y) # Floor Division -> Quotient
print(x % y) # Modulo operator -> giving the remainder
print(2 ** 3) # Exponentional Operator

z = -x 
print(z) # Negation
````

## Comparison or Relational Operators

````python
x = 10
y = 5 
z = 10.0

# check if x and y are equal 
print(x == y) # checks if two values are equal
print(x == z)
````

````python
x = 10
y = 5 
z = 10.0

print(x != y) # x is not equal to y

print("x is greater than y:", x > y)
print("x is smaller than y:", x < y)

print(x >= y)
print(x <= y)

print("Chaining two comparision", 10 < 50 < 100)
````

````python
## Assignment Operators

# Basic assignment
x = 5 # assigning value of 5 to x. so = is an assignment operator

x += 1 # x = x + 1
print(x)

x *= 10 # x = x * 10
print(x)

## Tuple unpacking
a, b = 2, 3
print(a + b)

student1, student2, student3 = "Ram", "Debashish", "Rajiv"
print(student1)
````

## Logical Operators

- and 
- or 
- not

````python
x = 5
y = 2 

print(x > 3 and y < 3)
print(x > 3 and y < 1)

print(x > 3 or y < 1)

print(not x > 3)
````

## Membership Operators

using python ```in``` and ```not in```

````python
name = "Rajiv"

print("j" in name)
print("p" in name)

print("p" not in name)
````

## Identity Operators
    - used to check if two items are the same object
    

````python
a = "Ramesh"
b = "Suresh"
c = a
d = "Ramesh"

print(a == d) # equality
print(a is d) # identity


print(a is b)

print(c is a)
````
